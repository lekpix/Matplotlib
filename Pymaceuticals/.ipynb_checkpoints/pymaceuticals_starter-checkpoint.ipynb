```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
print(mouse_metadata)
print(study_results)
# Combine the data into a single dataset
mouse_complete_data=pd.merge(mouse_metadata,study_results,on='Mouse ID',how='outer')
# Display the data table for preview
mouse_complete_data
```

```python
# Checking the number of mice.
mice_count=len(mouse_complete_data['Mouse ID'].unique())
print(f'There are {mice_count} mice')
```

```python
#mouse_complete_data.loc[mouse_complete_data["Mouse ID"]=='g989']
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_complete_data.loc[mouse_complete_data.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID']
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice=mouse_complete_data.loc[mouse_complete_data.duplicated(subset=['Mouse ID', 'Timepoint',])]
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleanMice_data=mouse_complete_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='first')
cleanMice_data
```

```python
# Checking the number of mice in the clean DataFrame
cleanMice_data['Mouse ID'].unique().shape[0]
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
regimen_mean=round(cleanMice_data.groupby('Drug Regimen').mean()['Tumor Volume (mm3)'],2)
regimen_mean
regimen_median=round(cleanMice_data.groupby('Drug Regimen').median()['Tumor Volume (mm3)'],2)
regimen_median
regimen_variance=round(cleanMice_data.groupby('Drug Regimen').var()['Tumor Volume (mm3)'],2)
regimen_variance
regimen_SD=round(cleanMice_data.groupby('Drug Regimen').std()['Tumor Volume (mm3)'],2)
regimen_SD
regimen_SEM=round(cleanMice_data.groupby('Drug Regimen').sem()['Tumor Volume (mm3)'],2)
regimen_SEM

# Assemble the resulting series into a single summary dataframe.
pd.concat([regimen_mean,regimen_median,regimen_variance,regimen_SD,regimen_SEM],axis=1)
Summary_regimen=pd.DataFrame({'Mean':regimen_mean,'Median':regimen_median,'Variance':regimen_variance,'SD':regimen_SD,'SEM':regimen_SEM})
Summary_regimen
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
Capomulin_df = cleanMice_data.loc[cleanMice_data["Drug Regimen"] == "Capomulin",:]
Capomulin_df
# Start by getting the last (greatest) timepoint for each mouse
Capomulin_last = Capomulin_df.groupby('Mouse ID').max()['Timepoint']
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
Capomulin_vol = pd.DataFrame(Capomulin_last)
Capomulin_merge = pd.merge(Capomulin_vol, cleanMice_data, on=("Mouse ID","Timepoint"),how="left")
```

```python
Ramicane_df = cleanMice_data.loc[cleanMice_data["Drug Regimen"] == "Ramicane", :]
Ramicane_vol = pd.DataFrame(Ramicane_df.groupby('Mouse ID').max()['Timepoint'])
Ramicane_merge = pd.merge(Ramicane_vol, cleanMice_data, on=("Mouse ID","Timepoint"),how="left")
```

```python
Infubinol_df = cleanMice_data.loc[cleanMice_data["Drug Regimen"] == "Infubinol", :]
Infubinol_vol=pd.DataFrame(Infubinol_df.groupby('Mouse ID').max()['Timepoint'])
Infubinol_merge=pd.merge(Infubinol_vol,cleanMice_data,on=("Mouse ID","Timepoint"),how='left')
```

```python
Ceftamin_df = cleanMice_data.loc[cleanMice_data["Drug Regimen"] == "Ceftamin", :]
Ceftamin_vol=pd.DataFrame(Ceftamin_df.groupby('Mouse ID').max()['Timepoint'])
Ceftamin_merge=pd.merge(Ceftamin_vol,cleanMice_data,on=('Mouse ID',"Timepoint"),how='left')
```

```python
Capomulin_tumors = Capomulin_merge["Tumor Volume (mm3)"]
quartiles =Capomulin_tumors.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
print(f"The lower quartile of Capomulin tumors: {lowerq}")
print(f"The upper quartile of Capomulin tumors: {upperq}")
print(f"The interquartile range of Capomulin tumors: {iqr}")
print(f"The median of Capomulin tumors: {quartiles[0.5]} ")
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr=round(st.pearsonr(avg_capm_vol_df['Weight (g)'],avg_capm_vol_df['Average Tumor Volume'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {corr}")
```

```python
x_values = avg_capm_vol_df['Weight (g)']
y_values = avg_capm_vol_df['Average Tumor Volume']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
```

```python
# Add the linear regression equation and line to plot
plt.scatter(x_values,y_values,color="blue")
plt.plot(x_values,regress_values,"r-")
plt.title('Regression Plot of Mouse Weight Versus Average Tumor Volume')
plt.xlabel('Weight(g)')
plt.ylabel('Average Tumor Volume')
plt.annotate(line_eq, xy=(15, 45),color='red')
plt.savefig('Output/ScatterplotRegression.png', bbox_inches = "tight")
plt.show()
```
